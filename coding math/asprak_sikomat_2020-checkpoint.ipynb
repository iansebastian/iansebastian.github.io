````python
## Anda juga dapat menulis kode di sini.


def ratio(i):
    """
    Menghitung hasil a_{i+1}/a{i}
    Args:
        i : int
            menyatakan bilangan ke-berapa yang akan ditampilkan
    Returns:
        float
            bilangan ke-i
    """
    if i == 1 or i == 2:
        return 1
    else:
        return 1/(1/ratio(i-1) + 1/ratio(i-2))
````
