<h2>Drawing a Bifurcation Diagram</h2>


I learned <a href="https://ipython-books.github.io/121-plotting-the-bifurcation-diagram-of-a-chaotic-dynamical-system/">here</a>.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Logistic model:
$f(x) = rx(1-x)$

```python
def logistic_model(r, x):
    return r * x * (1-x)
```

```python
plt.style.use("seaborn-ticks")
x = np.linspace(0, 1)
_ = plt.plot(x, logistic_model(2, x))
plt.show()
```

$x^{(r)}_{n+1} = f(x^{(r)}_{n})$

```python
def plot_model(r=2, x0=0.5, reps=10, ax=None):
    _ = ax.plot(np.linspace(0, 1), logistic_model(r, np.linspace(0, 1)), 'k-', lw=1)
    _ = ax.plot([0, 1], [0, 1], 'k-', lw=1)

    x = x0
    for i in range(reps):
        y = logistic_model(r, x)

        _ = ax.plot([x, x], [x, y], 'r', lw=1)
        _ = ax.plot([x, y], [y, y], 'r', lw=1)
        _ = ax.plot(x, y, 'ok', markersize=4,
                    alpha=(i+1)/reps)
        x = y
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    return ax
```

```python
plt.style.use("default")

fig, ax = plt.subplots(3, 1, figsize=(4, 18))

for i in range(3):
    _ = plot_model(r=2+i*0.5, x0=.1, ax=ax[i])
    ax[i].set_title(f"r={2+i*0.5 :.1f}; $x_0$=0.1")

plt.show()
```

Lyapunov exponent, but we won't count that now...

$$\lambda(r) = \lim_{x \to \infty} \frac{1}{n} \sum^{n-1}_{i=0}\log \left| \frac{df_r}{dx} \left( x_{i}^{(r)} \right) \right|$$

```python
iters=10000
last=100

x = 1e-5 * np.ones(iters)
r = np.linspace(2.5, 4, iters)
```

```python
plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(12,8))

for i in range(iters):
    x = logistic_model(r, x)
    if i >= iters-100:
        _ = ax.plot(r, x, ',r', alpha=.06)
_ = plt.margins(0)

plt.savefig("bifurcation.png")
_ = plt.show()
```

```python
import matplotlib.animation as animation

x = 1e-5 * np.ones(iters)
r = np.linspace(2.5, 4, iters)

x_plot = np.empty(0,)
r_plot = np.empty(0,)
for i in range(iters):
    x = logistic_model(r, x)
    if i >= iters-100:
        r_plot = np.append(r_plot, np.linspace(2.5, 4, iters))
        x_plot = np.append(x_plot, x)

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(12,8))

l, = ax.plot([], [], ',r', alpha=.2)

def init():
    _ = l.set_xdata([])
    _ = l.set_ydata([])
    return l,


def update_line(i, line, r_plot, x_plot):
    _ = line.set_xdata(np.linspace(2.5, 4, i))
    _ = line.set_ydata(x_plot[r_plot <= 2.5 + 1.5*i/10000])
    return line,


_ = ax.set_xlim(2.5, 4)

bifu_anim = animation.FuncAnimation(fig, update_line, fargs=(l, r_plot, x_plot),
                                    interval=5, blit=True, init_func=init)
```

```python
from IPython.display import HTML
from matplotlib.animation import FFMpegWriter

writer = FFMpegWriter(fps=60, metadata=dict(artist='Ian Sebastian'), bitrate=1800)
bifu_anim.save("bifurcation_diag.mp4", writer=writer)
print("Video import done.")
```
